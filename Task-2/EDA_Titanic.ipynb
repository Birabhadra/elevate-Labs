```python
# 1. Import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 2. Load the dataset
df = pd.read_csv("Titanic-Dataset.csv")  # make sure the dataset is in the same folder
df.head()
```

```python
# 3. Summary Statistics
print(df.describe(include='all'))
print(df.info())
```

```python
# 4. Handling missing values (if not done already)
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

#df.drop(columns=['Cabin'], inplace=True)  # optional if too many missing values
```

```python
# Step 5: Histograms for numeric columns
df.hist(figsize=(12, 8), bins=20, edgecolor='black')
plt.tight_layout()
plt.show()
```

```python
# Step 6: Boxplots for numeric features
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['Age'])
plt.title("Boxplot of Age")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['Fare'])
plt.title("Boxplot of Fare")
plt.show()
```

```python
# Step 7: Correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Step 8: Pairplot for selected features
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived', palette='husl')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

```python
# Step 9: Insights from visualizations
# - Passengers in 1st class were more likely to survive.
# - Fare seems positively skewed with some high-value outliers.
# - Younger passengers had higher survival rates.
```
